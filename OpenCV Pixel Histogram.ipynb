## Histogram pixel distributions

```python
# Import packages

import cv2 as cv
import matplotlib.pyplot as plt

# Read the image 

img = cv.imread('/Users/tradk/dogs-vs-cats/train/cat.1.jpg')
```

```python
# Grayscale image

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Gray Scale hist
gray_hist = cv.calcHist([gray],[0],None, [256],[0,256])


# Use Matplotlib to plot our histogram. 

plt.figure()
plt.title("Histogram")
plt.xlabel('Bins')
plt.ylabel("# of Pixals")
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()
```

```python
# Now we do a color histogram for each channel. 

plt.figure()
plt.title("Histogram")
plt.xlabel('Bins')
plt.ylabel("# of Pixals")
colors = ('b','g','r')
for i,col in enumerate(colors):
    hist = cv.calcHist([img],[i],None, [256],[0,256])
    plt.plot(hist,color = col)
    plt.xlim([0,256])
    
plt.show()
```
